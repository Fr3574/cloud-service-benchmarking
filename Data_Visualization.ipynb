# Data Visualization of the results

## Import packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

## Read the data into pandas dataframe

```python
def read_data(experiment_dir):
    # Get the settings
    sut, mode, trace_length, increment_interval = experiment_dir.split("/")[-1].split("_")[:4]
    sut_data = pd.read_csv(os.path.join(experiment_dir, f"output_{sut}.csv"))
    if mode == "horizontal":
        client_data = pd.read_csv(os.path.join(experiment_dir, f"output_{sut}_benchmark.csv"))
    else:
        dfs = []
        for file in [f for f in os.listdir(experiment_dir)]:
            if file == f"output_{sut}.csv":
                continue
            dfs.append(pd.read_csv(os.path.join(experiment_dir, file)))
        df = pd.concat(dfs)
        client_data = df.groupby("read").mean()
        client_data["traces"] = client_data["traces"].round().astype(int)
        client_data = client_data.reset_index()
        client_data["read"] = client_data["read"].str.split("m=").str[0].str.strip()
    # Merge dataframes on timestamp
    sut_data["timestamp"] = pd.to_datetime(sut_data["read"].str.replace(" UTC", ""), format="%Y-%m-%d %H:%M:%S.%f %z")
    client_data["timestamp"] = pd.to_datetime(client_data["read"].str.replace(" UTC", ""), format="%Y-%m-%d %H:%M:%S.%f %z")
    return pd.merge_asof(sut_data, client_data, on="timestamp", allow_exact_matches=False)
```

## Process data 

```python
def process_data(df):
    # Normalize to 100 %
    df["memory_percent"] = df["memory_percent"].str.replace("%","").astype(float) / 2
    df["cpu_usage"] = df["cpu_usage"] / 2
    # Split values using " / " as separator
    split_values_net = df["net_io"].str.split(" / ")
    split_values_block = df["block_io"].str.split(" / ")
    # Create two new columns with extracted values and remove the trailing "MB"
    df["network_input"] = split_values_net.str.get(0).str.replace("MB","")
    df["network_output"] = split_values_net.str.get(1).str.replace("MB","")
    df["block_input"] = split_values_block.str.get(0).str.replace("MB","")
    df["block_output"] = split_values_block.str.get(1).str.replace("MB","")
    df[["network_input", "network_output", "block_input", "block_output"]] = df[["network_input", "network_output", "block_input", "block_output"]].astype(float)
    # Add a second difference column 
    df[["network_input_diff", "network_output_diff", "block_input_diff", "block_output_diff"]] = df[["network_input", "network_output", "block_input", "block_output"]].diff()
    # Remove first row due to NaN entries (caused by the merge on the timestamp)
    df = df.drop(index=0)
    # Strip whitespaces in column names
    df.columns = df.columns.str.strip()
    # Remove duplicate entries (will cause errors in plotting)
    if "number_containers" in df.columns:
        df = df.drop_duplicates(subset="number_containers", keep="first")
    if "traces" in df.columns:
        df = df.drop_duplicates(subset="traces", keep="first")
    # Add an index column
    df["index"] = range(len(df))
    # Delete unuseful columns
    df = df.drop(columns=["read_x", "read_y", "net_io", "block_io"])
    return df
```

## Create labels for each column

```python
column_labels = {
    "number_containers": "Number of workers",
    "traces": "Number of parallel traces",
    "cpu_usage": "CPU Usage (%)",
    "memory_percent": "Memory Usage (%)",
    "memory_usage": "Memory Usage (MB)",
    "memory_limit": "Memory Limit (MB)",
    "pids": "Number of processes",
    "restart_count": "Cumulative number of restarts",
    "network_input": "Cumulative Data received from client (MB)",
    "network_output": "Cumulative Data written to storage (MB)",
    "network_input_diff": "Data received from client (MB)",
    "network_output_diff": "Data written to storage (MB)",
    "block_input": "Cumulative Data received from host system (MB)",
    "block_output": "Cumulative Data written to host system (MB)",
    "block_input_diff": "Data received from host system (MB)",
    "block_output_diff": "Data written to host system (MB)"
}
```

## Compare a metric of two SUTs against each other

```python
def compare_suts(x_column, y_column, df_A, title="", label_A="df_A", df_B=None, label_B="df_B"):
    fig, ax = plt.subplots()
    sns.lineplot(x=x_column, y=y_column, data=df_A, label=label_A)
    # add mean line
    ax.axhline(df_A[y_column].mean(numeric_only=True), color="#1f77b4", linestyle="--", alpha=0.5,
               label=r"$\mu={}$".format(round(df_A[y_column].mean(numeric_only=True),3)))
    if df_B is not None:
        sns.lineplot(x=x_column, y=y_column, data=df_B, label=label_B)
        ax.axhline(df_B[y_column].mean(numeric_only=True), color="#ff7f0e", linestyle="--", alpha=0.5,
                   label=r"$\mu={}$".format(round(df_B[y_column].mean(numeric_only=True),3)))
    # Add legend
    ax.legend()
    # Add labels
    ax.set_xlabel(column_labels[x_column])
    ax.set_ylabel(column_labels[y_column])
    # Set title
    ax.set_title(title)
    # Print the plot
    fig.show
```

## Compare two metrics of a sut

```python
def compare_metrics(x_column, df, y_column_A, title="", y_column_B=None):
    fig, ax = plt.subplots()
    sns.lineplot(x=x_column, y=y_column_A, data=df, ax=ax)
    # Add labels
    ax.set_xlabel(column_labels[x_column])
    ax.set_ylabel(column_labels[y_column_A], color="#1f77b4")
    if y_column_B is not None:
        ax2 = ax.twinx()
        sns.lineplot(x=x_column, y=y_column_B, data=df, ax=ax2, color="#ff7f0e")
        ax2.set_ylabel(column_labels[y_column_B], color="#ff7f0e")
    # Set title
    ax.set_title(title)
    # Print the plot
    fig.show
```

## Results

```python
jaeger_df_vertical = read_data("results/jaeger_vertical_1_0.01_2023-02-06_15:08:39")
tempo_30s_df_vertical = read_data("results/tempo_vertical_1_0.01_2023-02-09_00:58:04")
tempo_5m_df_vertical = read_data("results/tempo_vertical_1_0.01_2023-02-07_10:35:11")

jaeger_df_horizontal = read_data("results/jaeger_horizontal_0.005_3_2023-02-04_18:49:25")
tempo_30s_df_horizontal = read_data("results/tempo_horizontal_0.005_3_2023-02-07_18:43:12")
tempo_5m_df_horizontal = read_data("results/tempo_horizontal_0.005_3_2023-02-04_21:21:10")
```

```python
jaeger_df_vertical = process_data(jaeger_df_vertical)
tempo_30s_df_vertical = process_data(tempo_30s_df_vertical)
tempo_5m_df_vertical = process_data(tempo_5m_df_vertical)

jaeger_df_horizontal = process_data(jaeger_df_horizontal)
tempo_30s_df_horizontal = process_data(tempo_30s_df_horizontal)
tempo_5m_df_horizontal = process_data(tempo_5m_df_horizontal)
```

```python
compare_suts("number_containers", "memory_usage", tempo_30s_df_horizontal, "Comparison of different tempo block durations", "Tempo 30s", tempo_5m_df_horizontal, "Tempo 5m")
```

```python
compare_suts("number_containers", "cpu_usage", tempo_30s_df_horizontal, "CPU usage (horizontal)", "Tempo", jaeger_df_horizontal, "Jaeger")
compare_suts("number_containers", "memory_usage", tempo_30s_df_horizontal, "Memory consumption (horizontal)", "Tempo", jaeger_df_horizontal, "Jaeger")
compare_suts("number_containers", "network_output_diff", tempo_30s_df_horizontal, "Network output", "Tempo", jaeger_df_horizontal, "Jaeger")
```

```python
compare_suts("traces", "cpu_usage", tempo_30s_df_vertical, "CPU usage (vertical)", "Tempo", jaeger_df_vertical, "Jaeger")
compare_suts("traces", "memory_usage", tempo_30s_df_vertical, "Memory consumption (vertical)", "Tempo", jaeger_df_vertical, "Jaeger")
compare_suts("traces", "network_output_diff", tempo_30s_df_vertical, "Network output", "Tempo", jaeger_df_vertical, "Jaeger")
```

```python
compare_metrics("number_containers", tempo_30s_df_horizontal, "network_input_diff", "Tempo", "network_output_diff")
compare_metrics("number_containers", jaeger_df_horizontal, "network_input_diff", "Jaeger", "network_output_diff")
```

```python
compare_metrics("traces", tempo_30s_df_vertical, "network_input_diff", "Tempo 30s", "network_output_diff")
compare_metrics("traces", jaeger_df_vertical, "network_input_diff", "Jaeger", "network_output_diff")
```
